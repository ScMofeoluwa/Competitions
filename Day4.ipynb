```
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
from sklearn.datasets import load_breast_cancer
```

```
cancer = load_breast_cancer()
X = cancer.data
y = cancer.target
columns = cancer.feature_names
cancer_df = pd.DataFrame(X,columns=columns)
cancer_df['label'] = y
```

```
corr = cancer_df.corr()
```

```
sage = corr.label.abs().sort_values(ascending=False).head(11)
sage
```

```
top_10 = [i for i in sage.index]
top_10_pro = top_10[::-1]
top_10_pro
```

```
new_df = cancer_df[top_10_pro]
new_df
```

```
mage = cancer_df[top_10[1:6]]
mage
```

```
sns.pairplot(mage,diag_kind='hist')
```
