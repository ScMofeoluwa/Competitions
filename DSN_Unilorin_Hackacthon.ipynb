```
X = train_df.drop(['id','target'],axis=1)
y = train_df['target']
```

```
from sklearn.model_selection import train_test_split
```

```
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=1,test_size=0.4)
```

```
from catboost import CatBoostRegressor
```

```
cat = CatBoostRegressor(eval_metric='RMSE',random_state=1,random_strength=0,n_estimators=5000,
                        learning_rate=0.02,depth=12,l2_leaf_reg=10,
                        bagging_temperature=0.2,od_type='Iter',metric_period=75,od_wait=100)
```

```
cat.fit(X_train,y_train,
        eval_set=[(X_train,y_train),(X_test,y_test)],early_stopping_rounds=500,
        use_best_model=True,verbose=True)
```

```
cat.fit(X,y)
```

```
from sklearn.metrics import mean_squared_error as mse
```

```
pred = cat.predict(X_test)
np.sqrt(mse(y_test,pred))
```
